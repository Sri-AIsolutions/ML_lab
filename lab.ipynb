```python
# ML Lab 1 — Clean VS Code Version

import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics.pairwise import cosine_similarity
# Update this path to where your Excel file exists locally
file_path = r"D:\ML_LAB\ml_lab_data.xlsx"  # <-- CHANGE THIS

def mean_manual(data):
    return sum(data) / len(data)


def variance_manual(data):
    m = mean_manual(data)
    return sum((x - m) ** 2 for x in data) / len(data)


def jaccard_smc(v1, v2):
    f11 = np.sum((v1 == 1) & (v2 == 1))
    f00 = np.sum((v1 == 0) & (v2 == 0))
    f01 = np.sum((v1 == 0) & (v2 == 1))
    f10 = np.sum((v1 == 1) & (v2 == 0))

    jc = f11 / (f01 + f10 + f11) if (f01 + f10 + f11) else 0
    smc = (f11 + f00) / (f11 + f00 + f01 + f10)
    return jc, smc

purchase_df = pd.read_excel(file_path, sheet_name="Purchase data")

X = purchase_df[["Candies (#)", "Mangoes (Kg)", "Milk Packets (#)"]].values
y = purchase_df["Payment (Rs)"].values

rank = np.linalg.matrix_rank(X)
model = np.linalg.pinv(X) @ y

print("Rank:", rank)
print("Costs → Candies, Mangoes, Milk:", model)

purchase_df["Status"] = purchase_df["Payment (Rs)"].apply(lambda x: "RICH" if x > 200 else "POOR")

le = LabelEncoder()
y_encoded = le.fit_transform(purchase_df["Status"])

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y_encoded)

print("Predictions:", le.inverse_transform(knn.predict(X[:5])))

## 6️⃣ A3: IRCTC Stock Analysis

irctc_df = pd.read_excel(file_path, sheet_name="IRCTC Stock Price")
prices = irctc_df["Price"].dropna().values

print("Mean (Manual):", mean_manual(prices))
print("Mean (NumPy):", np.mean(prices))

loss_prob = len(irctc_df[irctc_df["Chg%"] < 0]) / len(irctc_df)
print("Loss Probability:", loss_prob)

plt.figure(figsize=(7,4))
sns.scatterplot(x="Day", y="Chg%", data=irctc_df)
plt.show()


## 7️ A4–A9: Thyroid Dataset Processing

thyroid_df = pd.read_excel(file_path, sheet_name="thyroid0387_UCI")

numeric_cols = ['age','TSH','T3','TT4','T4U','FTI','TBG']
for col in numeric_cols:
    thyroid_df[col] = pd.to_numeric(thyroid_df[col].replace('?', np.nan))

for col in numeric_cols:
    thyroid_df[col].fillna(thyroid_df[col].median(), inplace=True)

binary_cols = ['on thyroxine','query on thyroxine','sick','pregnant','thyroid surgery']
binary_df = thyroid_df[binary_cols].applymap(lambda x: 1 if x == 't' else 0)

jc, smc = jaccard_smc(binary_df.iloc[0].values, binary_df.iloc[1].values)
print("Jaccard:", jc, "SMC:", smc)

v1 = thyroid_df[numeric_cols].iloc[0].values.reshape(1,-1)
v2 = thyroid_df[numeric_cols].iloc[1].values.reshape(1,-1)

print("Cosine Similarity:", cosine_similarity(v1, v2)[0][0])

sim_matrix = cosine_similarity(binary_df.head(20))
sns.heatmap(sim_matrix, cmap="YlGnBu")
plt.show()

scaler = MinMaxScaler()
thyroid_df[numeric_cols] = scaler.fit_transform(thyroid_df[numeric_cols])
print(thyroid_df.head())


```
